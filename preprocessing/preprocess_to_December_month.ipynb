```python
import pandas as pd
import re

# Extract date from URLs
def extract_date(url):
    match = re.search(r'/(\d{4})/(\d{1,2})/(\d{1,2})/', url)
    if match:
        year = match.group(1)
        month = match.group(2).zfill(2)
        day = match.group(3).zfill(2)
        return f"{year}-{month}-{day}"
    return None

dfAJ['Date'] = dfAJ['URL'].apply(extract_date)
dfAJ.to_csv('news_with_dates.csv', index=False, encoding='utf-8-sig')
print("The file saved successfully")
```

```python
dfAJ.head()
```

```python
# Number of rows that don't belong to December - 2023
not_december_2023 = dfAJ[~dfAJ['Date'].str.startswith('2023-12', na=False)].shape[0]
print(f"Number of rows that don't belong to 2023 December is :   {not_december_2023}")
```

```python
# Just keep news that belong to December 2023
dfAJ = dfAJ[dfAJ['Date'].str.startswith('2023-12', na=False)]

# Reindex
dfAJ = dfAJ.reset_index(drop=True)

# Save to csv file
dfAJ.to_csv('AljazeeraNewsDecember_Cleaned.csv', index=False, encoding='utf-8-sig')
print("The file saved successfully")
```

### Extract news titles and details from the URLs 

```python
import pandas as pd
from newspaper import Article

df = pd.read_csv('AljazeeraNewsDecember_Cleaned.csv')  

def extract_article_info(url):
    try:
        article = Article(url, language='ar') 
        article.download()
        article.parse()
        return article.title, article.text
    except:
        return "", ""

df[['title', 'news_content']] = df['URL'].apply(lambda u: pd.Series(extract_article_info(u)))
df = df[df['title'] != ""].reset_index(drop=True)
df.to_csv('AJ_Articles.csv', index=False, encoding='utf-8-sig')
print("The file saved successfully")
```
